```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('dataset/titanic.csv')
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop(['PassengerId','Ticket','Fare','Cabin','Name'] , inplace=True , axis=1)
```

```python
df
```

```python
df['Survived'].fillna(method='ffill' , inplace=True)
```

```python
df['Age'].fillna(method='ffill' , inplace=True)
```

```python
df['Embarked'].fillna(method='ffill' , inplace=True)
```

```python
df
```

```python
x = df.iloc[:,1:]
x
```

```python
y = df.iloc[:,0]
y
```

```python
df.info()
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
tr = ColumnTransformer(transformers=[
    ('t2',OneHotEncoder(drop='first',sparse=False),['Sex','Embarked'])
] , remainder='passthrough')
```

```python
x1 = tr.fit_transform(x)
```

```python
x1
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x1 , y , test_size=0.2 ,random_state=2)
```

```python
x_train
```

```python
x_test.shape
```

```python
y_train
```

```python
y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier()
```

```python
dtc.fit(x_train,y_train)
```

```python
y_pred = dtc.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score , confusion_matrix
```

```python
print("Accuracy matrix =",accuracy_score(y_test,y_pred))
```

```python
pd.DataFrame(confusion_matrix(y_test , y_pred)) # if all the values are comes along the diagonal then accuracy is 100%
```

```python
result = pd.DataFrame()
result['Actual Label'] = y_test
result['Decision Tree Prediction'] = y_pred
```
